```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
from random import randint
from scipy import stats
```

# Create Data

```python
X = np.random.uniform(0, 300, 500)
Y = (X / 20) + np.random.uniform(0, 4, 500) + randint(0, 1)

plt.scatter(X, Y)
```

# Reshape Data

```python
X = X.reshape((500, 1))
Y = Y.reshape((500, 1))
print(X.shape, Y.shape)
```

# Estimate four values of m

```python
m_test_1 = 0.04
Y_pred_1 = X * m_test_1

m_test_2 = 0.05
Y_pred_2 = X * m_test_2

m_test_3 = 0.055
Y_pred_3 = X * m_test_3

m_test_4 = 0.07
Y_pred_4 = X * m_test_4
```

# Obtaining the correctly m by the linear least squares way

```python
m = np.matmul(inv(np.matmul(X.T, X)), np.matmul(X.T, Y))
m = m[0, 0]
print(m)

Y_pred = X * m
```

# Draw a scatter

```python
plt.scatter(X, Y, c='blue')

plt.plot(X, Y_pred_1, c='orange', lw=1)
plt.plot(X, Y_pred_2, c='purple', lw=1)
plt.plot(X, Y_pred_3, c='pink', lw=1)
plt.plot(X, Y_pred_4, c='brown', lw=1)

plt.plot(X, Y_pred, c='red', lw=3)
```

```python
X_list = []
for i in range(len(X)):
    X_list.append(X[i][0])
    
X_list
```

```python
Y_list = []
for i in range(len(Y)):
    Y_list.append(Y[i][0])
    
Y_list
```

# Obtaining fitted line with _scipy.stats.linregress_

```python
# slope, intercept, r_value, p_value, stderr
res = stats.linregress(X_list, Y_list)
```

```python
plt.plot(X, Y, 'o', label='original data')
plt.plot(X, res.intercept + res.slope * X, 'black', label='fitted line')
plt.legend()
plt.show()
```

# Result Graph

```python
plt.scatter(X, Y, c='blue', label='original data')

plt.plot(X, Y_pred_1, c='orange', lw=1, label='test 1')
plt.plot(X, Y_pred_2, c='purple', lw=1, label='test 2')
plt.plot(X, Y_pred_3, c='pink', lw=1, label='test 3')
plt.plot(X, Y_pred_4, c='brown', lw=1, label='test 4')

plt.plot(X, Y_pred, c='red', lw=3, label='manual lls line')

plt.plot(X, res.intercept + res.slope * X, 'black', lw=3, label='scipy lls line')

plt.legend()
plt.show()
```
