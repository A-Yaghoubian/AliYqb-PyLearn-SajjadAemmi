```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
```

گسسته -> classification (دسته بندی)

پیوسته -> regression (تخمین زنی)

```python
X = np.random.uniform(0, 100, 200)
Y = X + np.random.uniform(-10, 10, 200) * np.random.uniform(-4, 4, 200) + 15

plt.scatter(X, Y)
```

```python
X = X.reshape((200, 1))
X.shape
```

```python
Y = Y.reshape((200, 1))
Y.shape
```

```python
# m = (X.T X)^-1 * X.T Y 
m = np.matmul(inv(np.matmul(X.T, X)), np.matmul(X.T, Y))
m = m[0, 0]
m
```

```python
Y_pred = X * m
```

```python
plt.scatter(X, Y, c='blue')
plt.plot(X, Y_pred, c='red', lw=4)
```

```python
x_test = 70
```

```python
answer = x_test * m
answer
```
