```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../input/lowConFemale.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
equ = cv2.equalizeHist(img)
hist = cv2.calcHist([equ], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
alpha = 2 
beta = -150 
adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
hist = cv2.calcHist([adjusted], [0], None, [256], [0, 256])
plt.plot(hist)
cv2.imshow('q', adjusted)
```

```python
alpha2 = 1.5
beta2 = 0
new_image = cv2.addWeighted(img, alpha2, np.zeros(img.shape, img.dtype), 0, beta2)
hist = cv2.calcHist([new_image], [0], None, [256], [0, 256])
plt.plot(hist)
```
