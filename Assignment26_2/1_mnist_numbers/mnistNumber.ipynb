```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
mnist_img = cv2.imread("mnist.png")
mnist_img = cv2.cvtColor(mnist_img, cv2.COLOR_BGR2BGRA)
plt.imshow(mnist_img, cmap="gray")
```

```python
rows, cols, _ = mnist_img.shape
print(rows, cols)
```

```python
def extract_num(number):
    img = mnist_img[(rows // 10) * number : (rows // 10) * (number + 1), :]
    cv2.imwrite(f"{number}_mnist.png", img)
    return img
```

```python
for num in range(10):
    num_mnist_img = extract_num(num)
    num_rows, num_cols, _ = num_mnist_img.shape

    counter = 0
    for i in range(0, num_rows, num_rows//5):
        for j in range(0, num_cols, num_cols//100):
            small_final_img = num_mnist_img[i:i+(num_rows//5), j:j+(num_cols//100)]
            cv2.imwrite(f"numbers/{num}/{counter}.png", small_final_img)
            counter += 1        
        
```
