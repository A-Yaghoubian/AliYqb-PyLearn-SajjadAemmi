```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_BGR = cv2.imread('p1.jpg')
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)
img_HSV = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(img_HSV)
rows, cols, _ = img_BGR.shape
plt.imshow(img_RGB)
```

```python
black_background = np.zeros((rows, cols), dtype='uint8')
black_background_RGB = cv2.cvtColor(black_background, cv2.COLOR_GRAY2RGB)
plt.imshow(black_background_RGB)
```

```python
for i in range(rows):
    for j in range(cols):
        if 0 < H[i, j] < 20 and 35 < S[i, j] < 230 and 60 < V[i, j] < 250:
            black_background_RGB[i, j] = img_RGB[i, j]
```

```python
plt.imshow(black_background_RGB)
```

```python
cv2.imwrite('r1.jpg', cv2.cvtColor(black_background_RGB, cv2.COLOR_RGB2BGR))
```
