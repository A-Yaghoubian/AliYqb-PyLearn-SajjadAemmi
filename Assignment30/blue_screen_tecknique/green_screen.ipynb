```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_BGR = cv2.imread("SuperMan.jpg")
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)

print(img_RGB.shape)
plt.imshow(img_RGB)
```

```python
img_RGB = cv2.resize(img_RGB, (400, 300))
```

```python
img_HSV = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2HSV)

H, S, V = cv2.split(img_HSV)
```

```python
plt.imshow(H, cmap='gray')
```

```python
plt.imshow(S, cmap='gray')
```

```python
plt.imshow(V, cmap='gray')
```

```python
background_BGR = cv2.imread("sky.jpg")
background_RGB = cv2.cvtColor(background_BGR, cv2.COLOR_BGR2RGB)

plt.imshow(background_RGB)
```

```python
# background_RGB = cv2.resize(background_RGB, (2048, 1444))
background_RGB = cv2.resize(background_RGB, (400, 300))
```

```python
plt.imshow(background_RGB)
```

```python
rows, cols = H.shape

for i in range(rows):
    for j in range(cols):
        if (37 < H[i, j] < 75 and 40 < S[i, j] and V[i, j] > 81): # or (i < 50)
            img_RGB[i, j] = background_RGB[i, j]
           
```

```python
plt.imshow(img_RGB)
```

```python
cv2.imwrite('result.jpg', cv2.cvtColor(img_RGB, cv2.COLOR_RGB2BGR))
```
