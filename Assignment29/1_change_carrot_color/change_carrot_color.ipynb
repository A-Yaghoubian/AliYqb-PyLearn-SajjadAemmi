```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../input_data/Carrot.jpg')
img_beta = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_beta)
```

```python
R, G, B = cv2.split(img_beta)
result = cv2.merge((B, G, G))
```

```python
cv2.imwrite('green_carrot.jpg', result)
```
