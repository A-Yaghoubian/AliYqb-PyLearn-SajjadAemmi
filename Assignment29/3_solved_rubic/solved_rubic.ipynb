```python
import cv2
import matplotlib.pyplot as plt
```

```python
first_rubic = cv2.imread('../input_data/rubix.png')
plt.imshow(cv2.cvtColor(first_rubic, cv2.COLOR_BGR2RGB))
```

```python
rows, cols, _ = first_rubic.shape

for i in range(rows):
    for j in range(cols):
        b, g, r = first_rubic[i, j] 
        
        if b > 180 and g > 180 and r < 70:
            b = 0 
            g = 0
            r = 255
        
        if b > 180 and g < 70 and r > 180:
            b = 0 
            g = 255
            r = 0
        
        if b < 70 and g > 180 and r > 180:
            b = 255 
            g = 0
            r = 0
            
        first_rubic[i, j] = b, g, r
```

```python
cv2.imwrite('final.jpg', first_rubic)
```
