```python
import numpy as np
import matplotlib.pyplot as plt
```

melon width ~ 15 cm        std: 10

melon length ~ 25 cm       std: 10

melon weight ~ 2 kg        std: 1.5


Balloon width ~ 20 cm      std: 10

Balloon length ~ 30 cm     std: 10

Balloon weight ~ 0.018 kg   std: 0.01

```python
N = 100
std_w_h = 10
std_m = 1.5
std_b = 0.01

melon_width = np.random.normal(15, std_w_h, N)
melon_length = np.random.normal(25, std_w_h, N)
melon_weight = np.random.normal(2, std_m, N)

balloon_width = np.random.normal(20, std_w_h, N)
balloon_length = np.random.normal(30, std_w_h, N)
balloon_weight = np.random.normal(0.018, std_b, N)
```

```python
width = np.concatenate((melon_width, balloon_width))
length = np.concatenate((melon_length, balloon_length))
weight = np.concatenate((melon_weight, balloon_weight))
```

```python
X = np.array([width, length, weight])
```

```python
melon_label = np.zeros(N, dtype="int") # melon = 0
balloon_label = np.ones(N, dtype="int") # balloon = 1

Y = np.concatenate((melon_label, balloon_label))
```

```python
X.shape
```

```python
Y.shape
```

```python
X = X.T
X.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

xParameter = X[0:N, 0]
yParameter = X[0:N, 1]
zParameter = X[0:N, 2]

ax.scatter(xParameter, yParameter, zParameter)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

xParameter = X[N:, 0]
yParameter = X[N:, 1]
zParameter = X[N:, 2]

ax.scatter(xParameter, yParameter, zParameter)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

xParameter1 = X[0:N, 0]
yParameter1 = X[0:N, 1]
zParameter1 = X[0:N, 2]

xParameter2 = X[N:, 0]
yParameter2 = X[N:, 1]
zParameter2 = X[N:, 2]

ax.scatter(xParameter1, yParameter1, zParameter1, c='green', label='melon')
ax.scatter(xParameter2, yParameter2, zParameter2, c='red', marker='^', label='balloon')

ax.set_xlabel('X Label (WIDTH)')
ax.set_ylabel('Y Label (LENGTH)')
ax.set_zlabel('Z Label (WEIGHT)')

ax.set_title("Mellons & Balloons")

plt.show()
```
