```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Banana width = 4
# Banana length = 8
# Apple width = 6
# Apple length = 6

N = 1000
std = 0.5  # standard devision = ye taqribe

bananas_width = np.random.normal(4, std, N)
bananas_length = np.random.normal(8, std, N)

apples_width = np.random.normal(6, std, N)
apples_length = np.random.normal(6, std, N)

# mishe be jaye normal az uniform ham estefade kard vali halat haye mokhtalef ro baham barabar migire ke ma ino nemikhaym
# be in modeli ke pish raftim behesh migim tozi normal va nemodaresh zangoleii mishe
```

```python
width = np.concatenate((bananas_width, apples_width))
length = np.concatenate((bananas_length, apples_length))
```

```python
X = np.array([width, length])
```

```python
bananas_label = np.zeros(N, dtype="int") # banana = 0
apples_label = np.ones(N, dtype="int") # apple = 1

Y = np.concatenate((bananas_label, apples_label))
```
